```python
#6.What is the average time of a ride from pick up to drop off?
# to ensure pickup_ts and drop off are in datetime mode
ride_requests['pickup_ts'] = pd.to_datetime(ride_requests['pickup_ts'])
ride_requests['dropoff_ts'] = pd.to_datetime(ride_requests['dropoff_ts'])

# Filter out rows where either pickup_ts or dropoff_ts is missing (i.e., only completed rides)
completed_rides = ride_requests.dropna(subset=['pickup_ts', 'dropoff_ts'])

# Calculate ride duration (in seconds) for each ride
completed_rides['ride_duration'] = (completed_rides['dropoff_ts'] - completed_rides['pickup_ts']).dt.total_seconds()

# Calculate the average ride duration (in seconds)
average_ride_duration = completed_rides['ride_duration'].mean()

# Convert average ride duration to a more readable format (minutes or hours)
average_ride_duration_minutes = average_ride_duration / 60  # Convert seconds to minutes

print(f"Average ride duration: {average_ride_duration_minutes:.2f} minutes")
```

```python


# Download step
download = app_downloads.groupby('platform').agg(
    funnel_step=pd.NamedAgg(column='app_download_key', aggfunc=lambda x: 0),
    funnel_name=pd.NamedAgg(column='app_download_key', aggfunc=lambda x: 'download'),
    user_count=pd.NamedAgg(column='app_download_key', aggfunc='count'),
    ride_count=pd.NamedAgg(column='app_download_key', aggfunc=lambda x: 0)
).reset_index()

# Signup step
signup = app_downloads.merge(signups, left_on='app_download_key', right_on='session_id')
signup = signup.groupby('platform').agg(
    funnel_step=pd.NamedAgg(column='user_id', aggfunc=lambda x: 1),
    funnel_name=pd.NamedAgg(column='user_id', aggfunc=lambda x: 'signup'),
    user_count=pd.NamedAgg(column='user_id', aggfunc='count'),
    ride_count=pd.NamedAgg(column='user_id', aggfunc=lambda x: 0)
).reset_index()

# Ride requested step
ride_requested = app_downloads.merge(signups, left_on='app_download_key', right_on='session_id').merge(ride_requests, left_on='user_id', right_on='user_id', how='left')
ride_requested = ride_requested.groupby('platform').agg(
    funnel_step=pd.NamedAgg(column='user_id', aggfunc=lambda x: 2),
    funnel_name=pd.NamedAgg(column='user_id', aggfunc=lambda x: 'ride_requested'),
    user_count=pd.NamedAgg(column='user_id', aggfunc=pd.Series.nunique),
    ride_count=pd.NamedAgg(column='ride_id', aggfunc='count')
).reset_index()

# Ride accepted step
ride_accepted = app_downloads.merge(signups, left_on='app_download_key', right_on='session_id').merge(ride_requests, left_on='user_id', right_on='user_id', how='left')
ride_accepted = ride_accepted[ride_accepted['accept_ts'].notna()]
ride_accepted = ride_accepted.groupby('platform').agg(
    funnel_step=pd.NamedAgg(column='user_id', aggfunc=lambda x: 3),
    funnel_name=pd.NamedAgg(column='user_id', aggfunc=lambda x: 'ride_accepted'),
    user_count=pd.NamedAgg(column='user_id', aggfunc=pd.Series.nunique),
    ride_count=pd.NamedAgg(column='ride_id', aggfunc='count')
).reset_index()

# Ride completed step
ride_completed = app_downloads.merge(signups, left_on='app_download_key', right_on='session_id').merge(ride_requests, left_on='user_id', right_on='user_id')
ride_completed = ride_completed[ride_completed['dropoff_ts'].notna()]
ride_completed = ride_completed.groupby('platform').agg(
    funnel_step=pd.NamedAgg(column='user_id', aggfunc=lambda x: 4),
    funnel_name=pd.NamedAgg(column='user_id', aggfunc=lambda x: 'ride_completed'),
    user_count=pd.NamedAgg(column='user_id', aggfunc=pd.Series.nunique),
    ride_count=pd.NamedAgg(column='ride_id', aggfunc='count')
).reset_index()

# Ride charged step
ride_charged = app_downloads.merge(signups, left_on='app_download_key', right_on='session_id').merge(ride_requests, left_on='user_id', right_on='user_id').merge(transactions, left_on='ride_id', right_on='ride_id')
ride_charged = ride_charged[ride_charged['charge_status'] == 'Approved']
ride_charged = ride_charged.groupby('platform').agg(
    funnel_step=pd.NamedAgg(column='user_id', aggfunc=lambda x: 5),
    funnel_name=pd.NamedAgg(column='user_id', aggfunc=lambda x: 'ride_charged'),
    user_count=pd.NamedAgg(column='user_id', aggfunc=pd.Series.nunique),
    ride_count=pd.NamedAgg(column='ride_id', aggfunc='count')
).reset_index()

# Concatenating all the funnel steps into one DataFrame
funnel_data = pd.concat([download, signup, ride_requested, ride_accepted, ride_completed, ride_charged]).sort_values(['funnel_step', 'platform'])

# Display final funnel data
print(funnel_data)

```

```python
# Figure 2. Funnel plot of users showing % change

# Step 1: Define the data
data = {
    'funnel_step': [0, 1, 2, 3, 4, 5],
    'funnel_name': ['download', 'signup', 'ride_requested', 'ride_accepted', 'ride_completed', 'ride_charged'],
    'user_count': [1000, 800, 500, 300, 200, 150]
}

# Step 2: Create a DataFrame
df = pd.DataFrame(data)

# Step 3: Display the DataFrame (table)
print(df)

# Step 4: Group by funnel_step and funnel_name, summing the user_count
funnel_data = df.groupby(['funnel_step', 'funnel_name']).agg({'user_count': 'sum'}).reset_index()

# Step 5: Get the user_count value of the first level ("0")
first_level_value = funnel_data.loc[funnel_data['funnel_step'] == 0, 'user_count'].values[0]

# Step 6: Calculate the percentage change from the first level ("0")
funnel_data['percent_change_from_first'] = ((funnel_data['user_count'] - first_level_value) / first_level_value) * 100

# Step 7: Format the percentage changes for display in the graph
funnel_data['percent_change_text'] = funnel_data['percent_change_from_first'].apply(
    lambda x: f"{x:.2f}%" if x != 0 else "0%"
)

# Step 8: Create the funnel plot based on user_count
fig = go.Figure(go.Funnel(
    y=funnel_data['funnel_name'],  # Funnel stages
    x=funnel_data['user_count'],    # Total user count
    textinfo="value+text",          # Show value and custom text (variation)
    texttemplate='%{value:,} (%{customdata})',  # Format: value and percentage change
    customdata=funnel_data['percent_change_text'],  # Custom data for percentage
))

# Step 9: Configure the layout of the chart
fig.update_layout(
    title="Funnel Plot of Users with % Change from Download Level",
    xaxis_title="User Count",
    yaxis_title="Funnel Steps",
    autosize=False,
    width=800,
    height=600
)

# Step 10: Show the funnel plot
fig.show()
```

```python
#1. Using ride_request_table
# Sample data for "ride_requests" table
ride_request_data = {
    'ride_request_id': [1, 2, 3, 4, 5, 6, 7],
    'request_ts': [
        '2024-10-17 08:15:00',
        '2024-10-17 09:30:00',
        '2024-10-17 08:45:00',
        '2024-10-17 10:00:00',
        '2024-10-17 18:15:00',
        '2024-10-17 14:30:00',
        '2024-10-17 22:00:00'
    ],
}

# Create DataFrame for ride requests
ride_requests = pd.DataFrame(ride_request_data)

# Convert 'request_ts' to datetime
ride_requests['request_ts'] = pd.to_datetime(ride_requests['request_ts'])

# Extract hour from the request timestamp
ride_requests['request_hour'] = ride_requests['request_ts'].dt.hour

# Count the number of ride requests by hour
request_distribution = ride_requests.groupby('request_hour').size().reset_index(name='ride_request_count')

# Show the distribution of ride requests throughout the day
print(request_distribution)
```
